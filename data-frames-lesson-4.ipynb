## Basic operations with a data frame

The most common thing probably you want to do with data frame is subsetting. Subsetting means retreiving part of a large file that is of particular interest to us.

```r
# To do this, suppose we want to retreive the first 10 rows

stats[1:10,] # after column there's nothing which means we want all columns
```

```r
# If we want to retreive from the 3rd row to 9th row.

stats[3:9,]
```

```r
#If we want to retreive particular rows like for example the 4th row and the 100th row.

stats[c(4,100),]
```

One of the major benfit of subsetting is that it preserves the structure of the dataset. In the example above we can see that the row are numbered 4 and 100 and not 1 and 2.

```r
#Remember how [] works
```

```r
stats[1,] #let's say we only want to look at the first row.
```

Remember in matrix when we extract it drops the numbering 1,2,...and so on and it will drop the dimension turning the row in a vector. It doesn't happens in the data frame as we can see the number is still here.


```r
is.data.frame(stats[1,]) # To chec if its actually a dataframe. No need for (drop=F)
```

```r
stats[,1] # Here we can see its not a dataframe and is printed out in a form of a vector
```

```r
is.data.frame(stats[,1])
```

```r
#If we want to keep it as a data frame

stats[,1, drop=F] #Now the dimensions won't be dropped
```

Now we can see it is actually a data frame.


**Multiply Columns**

```r
# To multiply columns

stats$Birth.rate* stats$Internet.users
```

```r
# To add two columns

stats$Birth.rate + stats$Internet.users
```
