## The 'While' loop

```r
# the while loop checks the validity of statement after while 
while (FALSE) 
    {
    print ("Hello")
}

#it would not print anything as condition after while is FALSE
```

```r
#if we set the statement to always TRUE, we will get an infinite loop.

##while (TRUE )
   ## {
   ## print("Hello")
##}                        #not advised to run these infinite loops as it consumes cpu
```

```r
#using while loop to print all the numbers till 12

counter<- 1 #setting the counter to 1

while (counter < 12) #checking the condition of counter
    {
    print (counter)  #printing the value of counter
    
    counter <- counter + 1 #updating the value in the counter
    
}
```

# Law of large numbers



**x̄**:The average of value,suppose we toss a coin 100 times and get heads 43 times.The average will be 43/100=0.43
  
**n**: Size of Sample
  
**E(x)** : The actual (theoretical) value for a coin when tossed to return a head is 0.5

**Law of large number states that if the sample size gets increased to Infinity then the average value will coverge (be equal to) expected value of sample.**

# Normal distribution



**Suppose we take a million numbers generated by rnorm() function in R then by going through this graph of normal distribution we can understand that for those million numbers with a mean of 0 and standard deviation of 1, the probabilty of number falling between -1 and 1 is 68.2 %**

```r
# Let's write a R script that counts how many number fall between -1 and 1 and divide it by value of N 
```

```r
N <- 1000000  #sample size

counter <- 0 #initializing the counter with 0

for (i in rnorm(N)) {
    
    if (i > -1 & i < 1) {
        
        counter <- counter + 1
    }
}

answer <- counter / N
answer  # we can see that we are getting 0.68 which is theoretical value
```
